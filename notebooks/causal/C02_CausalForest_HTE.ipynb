---
## 1. Research Design
Identify $\tau(x) = E[Y(1)-Y(0)|X=x]$ nonparametrically.
Honest splitting: use separate samples for tree building and leaf estimation.


---
## 2. Mathematical Theory
**Pointwise asymptotic normality (Theorem 4.2):**
$$\frac{\hat{\tau}(x) - \tau(x)}{\sigma_n(x)} \xrightarrow{d} \mathcal{N}(0,1)$$
Variance from infinitesimal jackknife over trees.


---
## 3. Data


```python
from empirlab.causal import CausalForest
from empirlab.causal.datasets import make_hte_data
import numpy as np, pandas as pd, matplotlib.pyplot as plt

X, y, d, tau_true = make_hte_data(n=2000, p=10, seed=42)
print(f'True CATE range: [{tau_true.min():.2f}, {tau_true.max():.2f}]')
```

---
## 4. Estimation


```python
cf = CausalForest(n_estimators=1000).fit(X, y, d)
tau_hat = cf.predict(X)
lb, ub  = cf.confidence_interval(X)
print(cf.summary(X))
```

---
## 5. Results


```python
from scipy.stats import pearsonr
r, p = pearsonr(tau_true, tau_hat)
print(f'Correlation with true CATE: {r:.4f} (p={p:.4e})')
```

---
## 6. Visualization


```python
plt.figure(figsize=(6,5))
plt.scatter(tau_true, tau_hat, alpha=0.3, s=10)
plt.plot([tau_true.min(), tau_true.max()],[tau_true.min(), tau_true.max()],'r--')
plt.xlabel('True CATE'); plt.ylabel('Estimated CATE')
plt.title('Causal Forest: Estimated vs True CATE')
plt.tight_layout(); plt.show()
```
