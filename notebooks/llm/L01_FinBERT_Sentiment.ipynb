---
## 3. Data


```python
from empirlab.llm import FinSentiment
import pandas as pd

texts = [
    'Revenue grew 22% year-over-year, beating analyst expectations.',
    'The company issued a profit warning citing weak demand.',
    'Management remains cautiously optimistic about Q4 guidance.',
    'Operating costs increased significantly due to supply chain disruptions.',
    'The board declared a special dividend of $2.50 per share.',
]
```

---
## 4. Sentiment Scoring


```python
# Requires: pip install transformers torch
pipe = FinSentiment(model='finbert')
scores = pipe.score(texts)
df = pd.DataFrame({'text': texts, 'sentiment': scores})
print(df.to_string(index=False))
```

---
## 5. Results


```python
print(f'Mean sentiment: {df.sentiment.mean():.3f}')
print(f'Positive: {(df.sentiment > 0.2).sum()}')
print(f'Negative: {(df.sentiment < -0.2).sum()}')
```

---
## 6. Visualization


```python
import matplotlib.pyplot as plt
colors = ['green' if s > 0.2 else 'red' if s < -0.2 else 'gray'
          for s in df.sentiment]
plt.figure(figsize=(9, 3))
plt.barh(range(len(texts)), df.sentiment, color=colors)
plt.yticks(range(len(texts)), [t[:50]+'...' for t in texts], fontsize=8)
plt.axvline(0, color='black', lw=0.8)
plt.xlabel('Sentiment score'); plt.title('FinBERT Sentiment')
plt.tight_layout(); plt.show()
```
